```python
#добавляет новый столбец Spam в DataFrame data, который будет содержать бинарные значения, указывающие, является ли категория сообщения спамом.
data['Spam']=data['Category'].apply(lambda x:1 if x=='spam' else 0) # Лямбда-функция проверяет, равен ли элемент строке 'spam'. Если да, то возвращается 1, в противном случае — 0.
#Это позволяет создать бинарный индикатор, где 1 означает, что сообщение является спамом, а 0 — что нет.
data.head(5)
```

```python
from sklearn.model_selection import train_test_split # Импортируем функцию train_test_split из модуля sklearn.model_selection для разделения данных на обучающую и тестовую выборки.
X_train,X_test,y_train,y_test=train_test_split(data.Message,data.Spam,test_size=0.25)
#X_train: Сообщения для обучения.
#X_test: Сообщения для тестирования.
#y_train: Метки (спам/не спам) для обучающего набора.
#y_test: Метки для тестового набора.
#data.Message: Это столбец с текстами сообщений, который будет использоваться как входные данные (функции).
#data.Spam: Это столбец с метками (0 или 1), который будет использоваться как целевая переменная (метки).
#test_size=0.25: Указывает, что 25% данных будут отведены для тестирования, а оставшиеся 75% — для обучения.
```

```python
#Этот класс преобразует коллекцию текстовых документов в числовую матрицу, где строки представляют собой документы,
# а столбцы — уникальные слова (или токены) из всех документов. Значения в матрице представляют собой количество вхождений каждого слова в соответствующем документе.
from sklearn.feature_extraction.text import CountVectorizer
```

**Naive Bayes Have three Classifier(Bernouli,Multinominal,Gaussian)
Here I use Multinominal Bayes Because here data in a discrete form
discrete data(e.g movie ratings ranging  1 to 5 as each rating will have certain frequency to represent)**

Наивный байесовский алгоритм имеет три классификатора (Бернулли, полиномиальный, гауссовский). Здесь я использую полиномиальный байесовский алгоритм, потому что здесь данные в дискретной форме (например, рейтинги фильмов в диапазоне от 1 до 5, поскольку каждый рейтинг будет иметь определенную частоту для представления).

```python
from sklearn.naive_bayes import MultinomialNB
# модель наивного байесовского классификатора, что позволит вам классифицировать новые документы на основе обученных данных.
```

```python
from sklearn.pipeline import Pipeline # позволяет объединять несколько шагов обработки данных и моделирования в один объект
clf=Pipeline([ # Создает объект Pipeline, который будет последовательно применять указанные шаги
    ('vectorizer',CountVectorizer()), # векторизация текстовых данных с помощью CountVectorizer
    ('nb',MultinomialNB()) #применение модели Наивного Байеса для классификации векторизованных данны
])
```

# Tarining The Model

```python
# Обучение модели
clf.fit(X_train,y_train)
# модель анализирует входные данные (X_train) и соответствующие метки (y_train). На основе этого анализа она настраивает свои внутренние параметры, чтобы научиться различать спам и не спам.
```

**Here I given Two email Two detect 1st One is looking good and the other one looking spam**

Здесь я дал два электронных письма. Два обнаружения. Первый из них выглядит хорошо, а другой выглядит как спам.

```python
emails=[
    'Sounds great! Are you home now?',
    'Will u meet ur dream partner soon? Is ur career off 2 a flyng start? 2 find out free, txt HORO followed by ur star sign, e. g. HORO ARIES'
]
```

**Predict Email**

```python
clf.predict(emails) # предсказание меток для новых данных
```

# Prediction Of Model

```python
clf.score(X_test,y_test) # оценка качества модели на тестовом наборе данных
```
